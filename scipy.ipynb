```python
import numpy as np
import scipy as sp
from scipy import integrate
from scipy import interpolate
import matplotlib.pyplot as plt
x = np.linspace(0, 4, 12)
y = np.cos(x**2 / 3+4)
print("Array using linspace():\n", x)
print("\nTaking trigonometric cosine of the given array:\n", y)
```

```python
plt.plot(x, y)
plt.grid(color = "hotpink", ls = '-.', lw = '1')
plt.title("2D interpolation")
plt.xlabel("Array using linspace()")
plt.ylabel("Trigonometric cosine of the linspace() array")
plt.plot(x, y, marker = 'd', ms = '15', mec = 'darkgreen', mfc = 'darkgreen')
plt.show()
```

```python
import numpy as np
import scipy as sp
from scipy import integrate
from scipy import interpolate
import matplotlib.pyplot as plt
f1 = sp.interpolate.interp1d(x, y, kind = 'linear')
f2 = sp.interpolate.interp1d(x, y, kind = 'cubic')
xnew = np.linspace(0, 4, 30)
plt.plot(x,y)
plt.grid(color = "magenta", ls = ":", lw = '1')
plt.title("1D interpolation")
plt.xlabel("Array using linspace()")
plt.ylabel("Trigonometric cosine of the linspace() array")
plt.plot(x, y, marker = 'H', ms = '15', mec = 'navy', mfc = 'navy')
plt.show()
```

```python
# SCIPY - INPUT & OUTPUT
import scipy.io as sio
import numpy as np
# To save a Matlab(MAT) file
vect = np.arange(10)
sio.savemat('array.mat', {'vect' : vect})
# To load the file
mat_file_content = sio.loadmat('array.mat')
print("Saved MATLAB file:\n", mat_file_content)
# Using the whosmat command to inspect the content of a MATLAB file without reading it into memory
import scipy.io as sio
mat_file_content = sio.whosmat('array.mat')
print("Saved MATLAB file using whosmat:\n", mat_file_content)
```

```python
#sobel
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import imageio
from scipy import misc,ndimage
im = np.zeros((256, 256))
im[64:-64, 64:-64] = 1
im[90:-90, 90:-90] = 2
im = ndimage.gaussian_filter(im, 8)
sx = ndimage.sobel(im, axis = 0, mode = 'constant')
sy = ndimage.sobel(im, axis = 1, mode = 'constant')
sob = np.hypot(sx, sy)
plt.imshow(sob)
plt.show()
```
